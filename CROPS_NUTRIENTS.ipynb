```python
fig = px.scatter(crops1, x='AvMoisture%', y='AvN%(dry)', color='CropCategory',
                 title='Moisture vs Nitrogen Content by Crop Category',
                 labels={'AvMoisture%': 'Moisture Percentage', 'AvN%(dry)': 'Nitrogen Content (%)'})
fig.show()
```

```python
fig = px.histogram(nutrients_melted, x='Crop', y='Value', color='Nutrient',
             title='Average Nutrient Content by Crop',
             labels={'Value': 'Nutrient Content', 'Crop': 'Crop'})
fig.update_layout(xaxis_title='Crop', yaxis_title='Nutrient Content')
fig.show()
```

```python
fig = px.histogram(crops1, x='AvMoisture%', y='AvYieldUnitWeight(lb)', color='CropCategory',
                 title='Moisture vs Yield by Crop Category',
                 labels={'AvMoisture%': 'Moisture Percentage', 'AvYieldUnitWeight(lb)': 'Yield (lb)'})
fig.show()
```

```python
category_counts = crops1['CropCategory'].value_counts()

fig = px.pie(names=category_counts.index, values=category_counts.values,
             title='Proportion of Crop Categories')
fig.show()
```

```python
import pandas as pd
import plotly.express as px
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

# Prepare data for clustering
features = crops1[['AvMoisture%','AvN%(dry)', 'AvP%(dry)', 'AvK%(dry)', 'AvYieldUnitWeight(lb)']]
features = features.replace('#DIV/0!', pd.NA)
features1 = features.fillna(0)
pca = PCA(n_components=2)
reduced_features = pca.fit_transform(features1)

# Perform clustering
kmeans = KMeans(n_clusters=3, random_state=0).fit(reduced_features)
crops1['Cluster'] = kmeans.labels_

# Create a DataFrame for plotting
plot_df = pd.DataFrame(reduced_features, columns=['PC1', 'PC2'])
plot_df['Cluster'] = crops1['Cluster']
plot_df['Crop'] = crops1['Crop']

fig = px.histogram(plot_df, x='PC1', y='PC2', color='Cluster',
                 title='PCA Cluster Analysis of Crops',
                 labels={'PC1': 'Principal Component 1', 'PC2': 'Principal Component 2'})
fig.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score


# Define features and target
X = features1[['AvMoisture%', 'AvN%(dry)', 'AvP%(dry)', 'AvK%(dry)']]  # Features
y = features1['AvYieldUnitWeight(lb)']# Target

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict and evaluate
y_pred = model.predict(X_test)

y_pred = model.predict(X_test)
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("R^2 Score:", r2_score(y_test, y_pred))
```
