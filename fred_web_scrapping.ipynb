```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://fred.stlouisfed.org/categories/32991'
page = requests.get(url)
print(page)
```

```python
soup = BeautifulSoup(page.content,'html.parser')
soup
```

```python
soup.findAll(attrs = {'class':'content'})
```

```python
results = soup.find_all( 'span',class_ = "panel-title panel-title-center text-muted pull-left col-xs-12 col-sm-11")
print(results)
```

```python
tags = soup.find_all('li')
print(tags)
```

```python
results = soup.find('body',class_ = "money-banking-finance subcategories")
print(results.text)
title = soup.find('span',class_ = "panel-title panel-title-center text-muted pull-left col-xs-12 col-sm-11") 
unit =  soup.find('a',class_ = "burger-fred-gtm")
tags =  soup.find('li')
print(f'the position : {title.text}')
print(f'the unit : {unit.text}')
print(f'the tags : {tags.text}')
```

```python
fred_data = [[url,results,title,unit,tags]]
print(fred_data)
```

```python
import pandas as pd
```

```python
fred_data1 = pd.DataFrame(fred_data,columns = ['url','results','title','unit','tags'])
fred_data1
```

```python
results = soup.find('body',class_ = "money-banking-finance subcategories")
for i in results(1,10):
  url = 'https://fred.stlouisfed.org/categories/32991'
  page = requests.get(url)
  if(page.status_code==200):
    print("data fetched successfully",i)
    soup = BeautifulSoup(page.content,"html.parser")
    
    title = soup.find('span',class_ = "panel-title panel-title-center text-muted pull-left col-xs-12 col-sm-11")
    unit = soup.find('a',class_ = "burger-fred-gtm")
    tags = soup.find('li')
    print(f'the title : {title.text}')
    print(f'the unit : {unit.text}')
    print(f'the tags : {tags.text}')
    print(f'the results : {results}')
    fred_data.insert(results,[title,unit,tags])
  else:
    print("the url is not found",i)
df = pd.DataFrame(fred_data,columns = ['url','units','title','tags','results'])
df.to_csv('fred.csv')
df
```
