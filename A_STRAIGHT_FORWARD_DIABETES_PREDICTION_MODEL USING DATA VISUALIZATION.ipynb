```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go


fig1 = px.scatter_matrix(diabetes, dimensions=['Pregnancies', 'SkinThickness', 'DiabetesPedigreeFunction'], color='Outcome',
                         title='Scatter Plot Matrix')
fig1.show()

# 3D scatter plot
fig2 = px.scatter_3d(diabetes, x='Pregnancies', y='SkinThickness', z='DiabetesPedigreeFunction', color='Outcome',
                     title='3D Scatter Plot of Pregnancies, Skin Thickness, and DPF')
fig2.show()

# Box plots
fig3 = go.Figure()

# Box plot for Pregnancies
fig3.add_trace(go.Box(y=diabetes['Pregnancies'][diabetes['Outcome'] == 0], name='Pregnancies (Non-Diabetic)', marker_color='blue'))
fig3.add_trace(go.Box(y=diabetes['Pregnancies'][diabetes['Outcome'] == 1], name='Pregnancies (Diabetic)', marker_color='red'))

# Box plot for Skin Thickness
fig3.add_trace(go.Box(y=diabetes['SkinThickness'][diabetes['Outcome'] == 0], name='Skin Thickness (Non-Diabetic)', marker_color='blue'))
fig3.add_trace(go.Box(y=diabetes['SkinThickness'][diabetes['Outcome'] == 1], name='Skin Thickness (Diabetic)', marker_color='red'))
fig3.add_trace(go.Box(y=diabetes['DiabetesPedigreeFunction'][diabetes['Outcome'] == 0], name='DPF (Non-Diabetic)', marker_color='blue'))
fig3.add_trace(go.Box(y=diabetes['DiabetesPedigreeFunction'][diabetes['Outcome'] == 1], name='DPF (Diabetic)', marker_color='red'))

fig3.update_layout(title='Box Plots of Pregnancies, Skin Thickness, and DPF by Diabetes Outcome',
                   yaxis_title='Value', boxmode='group')
fig3.show()
```
